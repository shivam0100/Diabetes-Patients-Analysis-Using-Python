# Data Visualization

```python
sns.countplot(x = "Outcome", data = df, palette = "cool")

#plt.show()
```

```python
sns.countplot(x = "Pregnancies", data = df, palette = "Accent")
```

```python
sns.boxplot(x = "Outcome", y = "BMI", data = df, whis = 2.5)
```

```python
sns.set(rc = {"figure.figsize":(19,5)})
sns.countplot(x = "Age", data = df, palette = "Spectral")
plt.grid(False)
plt.show()
```

```python
sns.scatterplot(x = "Glucose", y = "BloodPressure", data = df, color = "g")
plt.grid(False)
plt.show()
```

```python
sns.violinplot(x = "Outcome", y = "BMI", data = df, palette = "Accent")
plt.grid(False)
plt.show()
```
